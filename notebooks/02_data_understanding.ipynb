```python
df_2 = pd.read_csv('/Users/meetnakrani/Library/Mobile Documents/com~apple~CloudDocs/recommendation-engine-platform/data/raw/ml-32m/movies.csv')
print(f'first of all main information of datasets :{df_2.head()}, {df_2.info()}, {df_2.describe()}')
print(
    f'The missing values in movieId: {df_2["movieId"].isnull().sum()}\n',
    f'The percentage of missing values in movieId: {round(df_2["movieId"].isnull().sum() / len(df_2) * 100, 2)}%'
)
print(
    f'The missing values in title: {df_2["title"].isnull().sum()}\n',
    f'The percentage of missing values in title: {round(df_2["title"].isnull().sum() / len(df_2) * 100, 2)}%'
)
print(
    f'The missing values in genres: {df_2["genres"].isnull().sum()}\n',
    f'The percentage of missing values in genres: {round(df_2["genres"].isnull().sum() / len(df_2) * 100, 2)}%'
)
print(
    f'checking duplicates value in title : {df_2["title"].duplicated().sum()} \n'
    f'checking duplicates vlaues in geners : {df_2["genres"].duplicated().sum()}'
)
print(
    f'chekinh unique values in movieId: {df_2["movieId"].nunique()} \n'
    f'checking unique value in title : {df_2["title"].nunique()} \n'
    f'checking unique vlaues in geners : {df_2["genres"].nunique()}'
)

# sorting values 
df_2["title_length"] = df_2["title"].str.len()

top_10 = (
    df_2.sort_values("title_length", ascending=False)
    [["title", "title_length"]]
    .head(10)
)

print("Top 10 Longest Movie Titles:")
print(top_10)
longest_idx = df_2["title"].str.len().idxmax()

print(df_2.loc[[longest_idx]])

df_2["title_length"] = df_2["title"].str.len()

top_10 = (
    df_2.sort_values("title_length", ascending=True)
    [["title", "title_length"]]
    .head(10)
)

print("Top 10 sortest Movie Titles:")
print(top_10)

shorest_idx = df_2["title"].str.len().idxmin()
print(df_2.loc[shorest_idx])

#extract relasing year 

df_2["year"] = df_2["title"].str.extract(r"\((\d{4})\)")


df_2["year"] = pd.to_numeric(df_2["year"], errors="coerce")


latest_year = df_2["year"].max()
oldest_year = df_2["year"].min()

print(f"Latest movie year in dataset: {latest_year}")

print("\nMovies released in the latest year:")
print(
    df_2[df_2["year"] == latest_year]
    [["title", "year"]]

)

print(f"\nOldest movie year in dataset: {oldest_year}")

print("\nMovies released in the oldest year:")
print(
    df_2[df_2["year"] == oldest_year]
    [["title", "year"]]
    
)


print(
    f'The values combinaation  in genres: {df_2["genres"].value_counts().head(20)}')

#finding number of geners in movie

df_2["genres_count"] = (df_2["genres"].str.split('|').apply(len))
print(df_2[["title","genres_count"]])

print(
    df_2.sort_values(
        "genres_count",
        ascending=True
    )[["title", "genres", "genres_count"]].head(4))

print(
    df_2.sort_values(
        "genres_count",
        ascending=False
    )[["title", "genres", "genres_count"]].head(4))

df_2.memory_usage(deep=True)

print(df_2.memory_usage(
    deep=True
).sum()/1024**2)

summary = pd.DataFrame({
    "Data Type": df_2.dtypes,
    "Null Count": df_2.isnull().sum(),
    "Null %": round(df_2.isnull().sum()/len(df_2)*100,2),
    "Unique Values": df_2.nunique()
})

print(summary)
```

```python
df_3 = pd.read_csv('/Users/meetnakrani/Library/Mobile Documents/com~apple~CloudDocs/recommendation-engine-platform/data/raw/ml-32m/ratings.csv')
print(f'the info of data is : {df_3.info()}')
print(f'the first 5 data is: {df_3.head()}')
print(f'the shape of  data is: {df_3.shape}')
print(f' data describe : {df_3.describe()}')


summary = pd.DataFrame({
    "Data Type": df_3.dtypes,
    "Missing Count": df_3.isnull().sum(),
    "Missing %": round(df_3.isnull().sum()/len(df_3)*100,2),
    "Unique Values": df_3.nunique()
})

print(summary)

print(f'dublicate values: {df_3.duplicated().sum()}')

#now for user analysis

print(f'the number of user: {df_3["userId"].unique()}')
print(f'most active users : {df_3["userId"].value_counts().head(10)}')

ratings_per_user = df_3.groupby("userId")["rating"].count()
print(ratings_per_user.describe())


#movie analysis
print(f'the number of rated movie: {df_3["movieId"].unique()}')
print(f'most rated  movies : {df_3["movieId"].value_counts().head(10)}')

rating_per_movie = df_3.groupby("movieId")["rating"].count()
print(rating_per_movie.describe())

#rating analysis
df_3["rating"].value_counts().sort_index()
print(f'avreage rating is: {df_3["rating"].mean()}')
print(f'median rating is: {df_3["rating"].median()}')

heighest_rated_movie = (df_3.groupby("rating")["movieId"].count()
                        .sort_index(ascending=False))
print(heighest_rated_movie.head(5))
print(heighest_rated_movie.tail(5))

#timestamp analysis

df_3["data"] = pd.to_datetime(df_3['timestamp'], unit='s')
print(f'the data & time of timestamp : {df_3["data"]}')
print(df_3["data"].min())
print(df_3["data"].max())
df_3["year"] = df_3["data"].dt.year

df_3["year"].value_counts().sort_index()

#now to merge movies.csv files

merged = df_3.merge(df_2,on = "movieId", how="left")
print("most rated movie:\n")
print(merged["title"].value_counts().head(10))

merged.groupby("title")["rating"] \
      .mean() \
      .sort_values(ascending=False) \
      .head(10)
```
